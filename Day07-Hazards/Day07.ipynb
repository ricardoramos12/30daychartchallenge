```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/ricardoramos12/30daychartchallenge/refs/heads/main/Day07-Hazards/dengue.csv")
sns.set(style="whitegrid")
# Crear un gráfico de densidad para 'EDAD' por 'SEXO' (opcional)
plt.figure(figsize=(10, 6))
sns.kdeplot(data=data, x='EDAD', hue='SEXO', fill=True, palette=['#FF6347', '#4682B4'])
plt.title('Distribución de Edades por Sexo en el caso de Infectados por dengue (ene 2023 - may 2023)')
plt.xlabel('Edad')
plt.ylabel('Densidad')
plt.show()
```
