```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = 'https://github.com/ricardoramos12/30daychartchallenge/raw/refs/heads/main/Day21-GreenEnergy/green.csv'
data = pd.read_csv(url)
```

```python
data_chile = data[data['Entity'] == 'Chile'].copy()
data_chile.loc[:, 'Year'] = data_chile['Year'].astype(int)

# Filtrar entre 2010 y 2020
data_chile = data_chile[(data_chile['Year'] >= 2010) & (data_chile['Year'] <= 2020)]
sns.set(style="whitegrid")

# Crear el gráfico de líneas
plt.figure(figsize=(12, 6))
sns.lineplot(data=data_chile, x='Year', y='Electricity from renewables (TWh)', marker='o')

# Añadir los valores de cada punto en la línea
for index, row in data_chile.iterrows():
    plt.text(row['Year'], row['Electricity from renewables (TWh)'],
             f'{row["Electricity from renewables (TWh)"]:.2f}',
             ha='center', va='bottom', fontsize=10)

# Personalizar el gráfico
plt.title('Electricidad en TWh de Fuentes Renovables en Chile (2010-2020)', fontsize=16)
plt.xlabel('Año', fontsize=14)
plt.ylabel('Electricidad (TWh)', fontsize=14)
plt.xticks(rotation=45)
plt.gca().xaxis.get_major_locator().set_params(integer=True)
plt.tight_layout()
plt.show()
```
