```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = 'https://github.com/ricardoramos12/30daychartchallenge/raw/main/Day13-Family/family.xlsx'
data = pd.read_excel(url)

data.set_index('Provincia', inplace=True)
data_T = data.T
data_T.index = pd.to_datetime(data_T.index.astype(str))

# Crear la gráfica de líneas
plt.figure(figsize=(10, 6))
for province in data_T.columns:
    plt.plot(data_T.index.year, data_T[province], marker='o', label=province)

    # Añadir etiquetas con los valores en cada punto
    for x, y in zip(data_T.index.year, data_T[province]):
        plt.text(x, y, str(y), fontsize=9, ha='center', va='bottom')

# Configurar el gráfico
plt.title('Evolución de matrimonios por provincia (2012-2022)')
plt.xlabel('Año')
plt.ylabel('Número de Matrimonios')
plt.xticks(data_T.index.year, rotation=45)
plt.legend(title='Provincias')
plt.grid()
plt.tight_layout()
plt.show()
```
