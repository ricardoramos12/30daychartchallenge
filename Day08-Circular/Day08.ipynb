```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/ricardoramos12/30daychartchallenge/refs/heads/main/Day07-Hazards/dengue.csv")

# Generar un conteo de registros por sexo
sexo_counts = data['SEXO'].value_counts()

# Configuración del gráfico circular con cantidades
plt.figure(figsize=(8, 6))
plt.pie(sexo_counts,
        labels=sexo_counts.index,
        autopct='%1.1f%%',
        startangle=140,
        colors=sns.color_palette("pastel"))

plt.title("Distribución de infectados por dengue en base a su sexo - Perú (ene 2023 - may 2023)")

# Añadir leyenda con etiquetas
plt.legend(title="Sexo", labels=[f"{label}: {count}" for label, count in zip(sexo_counts.index, sexo_counts)],
           loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.show()
```
